```python
from selenium import webdriver
import time
import re

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3) 

driver.get("https:/gs25.gsretail.com/gscvs/ko/products/event-goods")
time.sleep(2)

twotoone = driver.find_elements_by_css_selector('#TWO_TO_ONE')
for clickbtn in twotoone :
    clickbtn.click()
time.sleep(3)
goodsname = []
goodsprice = []
```

```python
while True:
    goods = driver.find_elements_by_css_selector('ul > li > div > p.tit')
    for good in goods:
        #print(good.text)
        if len(good.text) > 0:
            goodsname.append(good.text)

    prices = driver.find_elements_by_css_selector('ul > li > div > p.price > span')
    for price in prices:
        price= re.sub("[,원]", "", price.text)
        #print(price)
        if len(price) > 0:
            goodsprice.append(price)

    #print(goodsname,goodsprice)

    nextbtn = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div[3]/div/div/div[2]/div/a[3]')
    if not nextbtn.get_attribute('onclick') == 'goodsPageController.moveControl(1)':
        break
    nextbtn.click()
    time.sleep(1)

with open('output/gs25_twotoone.csv', "wt", encoding="utf-8") as f:
    f.write('goodsname,goodsprice\n')  
    for i in range(len(goodsname)):
        f.write(goodsname[i]+","+goodsprice[i]+'\n')  
```
